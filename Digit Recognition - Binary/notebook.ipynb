# Imports

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
```

# Data

```python
# I use lowercase y to denote a 1D tensor
X = np.load('data/X.npy')[:1000]
y = np.load('data/Y.npy')[:1000]
```

```python
print(f'The shape of X is {X.shape}')
print(f'The shape of y is {y.shape}')
```

Training set contains 1000 examples, where each input is a 1D array of 400 floating point numbers. <br>Each array contains 400 pixels. <br>These pixels can be rearranged into a 20 by 20 grid.

```python
print(f'The first training example is {X[0]}')
```

```python
print(f'which corresponds to the digit {y[0][0]}')
```

# Visualise Data

```python
# Create a figure shows 8*8 digits
num_images_per_row = 8
num_rows = 8
```

```python
# index the dataset at 64 random places
indices = np.random.randint(0,X.shape[0], size = num_images_per_row * num_rows)
X_sample = X[indices]
y_sample = y[indices]
```

```python
# draw figure
# notice how i reshape the data example and then transpose. Transpose swaps the rows and columns of our matrix so our digit the right way
fig, axes = plt.subplots(num_rows,num_images_per_row,figsize=(num_images_per_row*1,num_rows*1))
fig.tight_layout(pad=0.1)

for i, ax in enumerate(axes.flat):
    ax.imshow(X_sample[i].reshape(20,20).T,cmap='grey')
    ax.axis('off')
    ax.set_title(y_sample[i].item())
        
plt.show()
```

# Model

```python
model = Sequential([
    Input(shape=(400,)),
    Dense(25,activation='sigmoid',name='L1'),
    Dense(15,activation='sigmoid',name='L2'),
    Dense(1,activation='sigmoid',name='L3')],name='mymodel')
```

```python
model.summary()
```

```python
# before training
predictions = model.predict(X_sample)
yhats = [1 if predictions[i]>=0.5 else 0 for i in range(len(predictions))]

fig, axes = plt.subplots(num_rows,num_images_per_row,figsize=(num_images_per_row*1,num_rows*1))
fig.tight_layout(pad = 0.1, rect = [0.05,0,0.95,0.92])
fig.suptitle('Prediction, Truth', fontsize = 16)

for i, ax in enumerate(axes.flat):
    ax.imshow(X_sample[i].reshape(20,20).T,cmap='grey')
    ax.axis('off')
    ax.set_title(f'{yhats[i]},{y_sample[i].item()}')

plt.show()

print(f'{sum(yhats==y_sample.reshape(len(y_sample),))} out of {len(y_sample)} digits correctly predicted')
```

```python
# Define loss function
model.compile(
    loss=tf.keras.losses.BinaryCrossentropy(),
    optimizer=tf.keras.optimizers.Adam(0.001)
)

# Fit model to data
model.fit(X, y, epochs=20)
```

```python
# after training
predictions = model.predict(X_sample)
yhats = [1 if predictions[i]>=0.5 else 0 for i in range(len(predictions))]

fig, axes = plt.subplots(num_rows,num_images_per_row,figsize=(num_images_per_row*1,num_rows*1))
fig.tight_layout(pad = 0.1, rect = [0.05,0,0.95,0.92])
fig.suptitle('Prediction, Truth', fontsize = 16)

for i, ax in enumerate(axes.flat):
    ax.imshow(X_sample[i].reshape(20,20).T,cmap='grey')
    ax.axis('off')
    ax.set_title(f'{yhats[i]},{y_sample[i].item()}')

plt.show()

print(f'{sum(yhats==y_sample.reshape(len(y_sample),))} out of {len(y_sample)} digits correctly predicted')
```
